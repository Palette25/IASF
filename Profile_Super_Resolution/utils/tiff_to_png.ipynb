```python
from pathlib import Path

from PIL import Image


def convert_tiff_to_png(tiff_path, png_path=None, overwrite=False):
    """Convert a TIFF image to PNG.

    Args:
        tiff_path: Path to the input TIFF file.
        png_path: Path to the output PNG file. If None, uses the same stem with .png.
        overwrite: If True, overwrite an existing output file.

    Returns:
        Path to the saved PNG file.
    """

    tiff_path = Path(tiff_path)
    if png_path is None:
        png_path = tiff_path.with_suffix('.png')
    png_path = Path(png_path)

    if png_path.exists() and not overwrite:
        raise FileExistsError(f"Output file already exists: {png_path} (set overwrite=True to replace)")

    with Image.open(tiff_path) as img:
        # 如果是多页 TIFF，只取第一页
        if getattr(img, "n_frames", 1) > 1:
            img.seek(0)
        img = img.convert('L')  # 转为 8 位灰度，若需要 RGB 可改为 'RGB'
        img.save(png_path, format='PNG')

    return png_path
```

```python
# 使用示例
# 将当前目录下的 example.tiff 转为 example.png
convert_tiff_to_png('../Image__2025-08-13__11-53-03.tiff')
```
